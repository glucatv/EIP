# Esercizio per casa

Scrivere una funzione ricorsiva che prende in input una stringa e la stampa al contrario.

*Suggerimento*. Potrebbe essere utile dover eseguire più stampe sulla stessa riga di output. Allo scopo utilizzare, nella funzione `print`, il parametro `end=''` che impedisce alla funzione di andare a capo dopo aver stampato. Ad esempio

```
    print('stringa 1', end='')
    print('stringa 2')
```

Stampa `stringa 1stringa2`.

## Soluzione

````python
def apmats( a ):
    if len(a) <= 1:
        print(a, end='')
    else:
        apmats(a[1:])
        print(a[0], end='')
        
apmats('abcdefg')
````

## Esercizio

Scrivere una funzione che prenda in input un nome di una directory e restituisca un dizionario delle estensioni dei file nella directory. Ovvero un dizionario che associ ogni estensione alla lista dei file con quell'estensione.

### Soluzione

````python
def estensioni( dirname ):
    if os.path.isdir(dirname) == False:
        return None
    contenuto_cartella = os.listdir( dirname )
    diz_estensioni = {}
    for filename in contenuto_cartella:
        if os.path.isfile(filename):
            lista_split = filename.split('.')
            if len(lista_split) == 1: # file senza estensione
                ext = None
            else:
                ext = lista_split[-1]
            if ext in diz_estensioni:
                diz_estensioni[ext].append(filename)
            else:
                diz_estensioni[ext] = [filename]
    return diz_estensioni
````

````python
def browse_dir( cartella ):
    contenuto_cartella = os.listdir( cartella )
    for filename in contenuto_cartella:
        new_path = os.path.join(cartella, filename)
        if os.path.isdir(new_path):
            browse_dir(new_path)
        else:
            print(new_path)
            
browse_dir('root')
````

````python
def browse_dir( cartella, livello = 0 ):
    contenuto_cartella = os.listdir( cartella )
    for filename in contenuto_cartella:
        new_path = os.path.join(cartella, filename)
        if os.path.isdir(new_path):
            browse_dir(new_path, livello+1)
        else:
            print(new_path, livello)
            
browse_dir('root')
````

````python
def find_in_dir( cartella, chiave ):
    '''
    Assunzione: cartella è il nome di una directory nella cartella di lavoro
        chiave è una stringa
    Stampa i nomi di tutti i file di testo in cartella e in tutte le sue
        sottocartelle che contengono la stringa chiave
    '''
    contenuto_cartella = os.listdir( cartella )
    for filename in contenuto_cartella:
        new_path = os.path.join(cartella, filename)
        if os.path.isdir(new_path):
            find_in_dir(new_path, chiave)
        else:
            est = new_path.split('.')[-1]
            if est in ['txt', 'py', 'ipynb']:
                fp = open(new_path)
                for line in fp.readlines():
                    if chiave in line:
                        print(new_path)
                        break
                fp.close()
````
